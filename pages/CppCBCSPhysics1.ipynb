### Structure of a C++ Program:
The C++ program has following components:
* Comments: These are used for better undertaking of program. Comments start with symbol //.
* `# include`: This is a directive to the compiler to include the contents of the file enclosed in < > into the source file.
* Function: A function is identified by a function name and a function body. All programs must have a function called main (). main() is int in standard c++. void main() is non-standard. The execution of a program starts with the main function. The keyword void along with the function name signifies that the function does no return any value. Every function must have a pair of braces ({}). Braces delimit a block of program statements.
* cout: This output statement causes the enclosed "text" to be displayed on screen.
* Data Types: Char, int and float are fundamental data type having number of bytes one, four and four.
* Variables: Variables are created using keyword int, char and float.
  * Integer variable: These are used to store whole numbers.
  * Character variable: These are used to store character constrats like A', 'B', 'a', b etc. character constants are enclosed with single quotes.
  * Strings: To store more than one character, a string has to be declared . For example cha c Word [20]: This statement allocated 20 contiguous bytes to hold a shing. The last byte is used to store the string terminator (0), known as NULL character. Memory locations for charactd variable and integer variable is one byte in size.
* Operators:
Arithmetic operators :+-* / and % the last one % is module operator to get the remainder in division.
* Arithmatic Assignment Operators (= ):
ix = ix + 2: is same as ix + = 2: This increment the value of variable ix by 2. The other are -=, * =, /=, % =.
* Relational operator:
This is same as in C language. They are >,<==,!=,>=, <= etc.
* Logical Operator:
C++ provides AND (&&), OR (I), NOT (!) logical operators
* Increment Operator: (++ or --) An increment operator can be used as ++ ix in place of ix = ix +1 or ix += 1.

An Armstrong number is a number that is equal to the sum of its own digits, each raised to the power of the total number of digits. All single digit numbers are Armstrong numbers. For two digit numbers, for an Armstrong number $ab=a^2+b^2$. There is no such number! For 3 digits, $abc=a^3+b^3+c^3$. For example, 153 is an Armstrong number because it has three digits, and $ 1^{3}+5^{3}+3^{3}=1+125+27=153$.

```python
%%writefile p4armstrong.cpp
//ARMSTRONG NUMBERS BETWEEN 1 TO N
# include <iostream>
# include <math.h>
int main ()
{
int n, n1, digit;
std::cout <<"INPUT\nEnter the number n ";
std::cin >> n;
std::cout <<"OUTPUT \n";
for (int i=1; i <=n; i ++)
{
  int y=i;
  int sum=0;
  digit =0;
  do
  {
    digit++;
    y=y/10;
   }while (y>0);
  y=i;
  do
  {
    n1=y% 10;
    sum=sum + pow(n1, digit);
    y=y/10;
   }while (y>0);
  if (i== sum) std::cout<< i<<" is a armstrong number.\n";
}
}
```

```python
%%writefile p5lcmgcd.cpp
// GCD AND LCM OF TWO NUMBERS USING OOP
# include <iostream>
using namespace std;

class gl
{
int m, n;
int gcdlcm () ; // a private function
public:
void getdata ()
{
  int p, q;
  cout<<" Enter both the numbers ";
  cin >> p >> q;
  m=p;
  n=q;
}
void show ()
{
  cout<<"GCD of the given two numbers is "<< gcdlcm ();
  cout<< "\n\n and LCM is "<< (m*n) /gcdlcm ();
}
};

int gl:: gcdlcm ()
{
int p, q;
p=m; q=n;
while (p!=q)
{
// Euclid's algorithm for GCD
if (p>q) p=p-q;
else q=q-p;
}
return p;
}


int main ()
{
  gl obj;
  cout<<"INPUT:\n";
  obj.getdata ();
  cout<<"OUTPUT\n";
  obj.show();
}
```
